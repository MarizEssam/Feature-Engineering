```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# List of columns to encode
cat_cols = ['Airline', 'Source', 'Destination', 'Additional_Info']

# Create encoder
ohe = OneHotEncoder(sparse_output=False)  # return dense array

for col in cat_cols:
    # Fit and transform the column
    encoded = ohe.fit_transform(df_final[[col]])

    # Create DataFrame with proper column names
    encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out([col]))

    # Reset index to align with df_final
    encoded_df.index = df_final.index

    # Concatenate with original DataFrame
    df_final = pd.concat([df_final, encoded_df], axis=1)

    # Drop original column
    df_final.drop(col, axis=1, inplace=True)

# df_final now has all categorical columns one-hot encoded
```
